# Demonstration on the polynomial Wigner Ville with the non-uniform FFT

```julia
using Winston;
using Color;
using PyPlot
import DSP
```

```julia
include("../juwvid.jl")
```

```julia
# multi component data
nsample=512
t,x=sampledata.genmultifm623x(nsample);
Winston.plot(t,x)
```

### polynomial WV with FFT

```julia
# the resolution is poor  
trfpo=polywv.tfrpowv(x,NaN,NaN,NaN,2,2);
imagesc(trfpo[80:-1:20,:])
```

### with NuFFT

```julia
# now, the resolution is fine  
fin=collect(linspace(20,120,1024));
trfpo=polywv.tfrpowv(x,NaN,NaN,fin,2,2);
imagesc(trfpo[end/2:-1:1,:])
```

### the Pseudo Wigner Ville

```julia
# you can see the severe cross talk between two components. 
z=DSP.Util.hilbert(x);
tfrpf=cohenclass.tfrpwv(z,NaN,NaN,2*fin,NaN,NaN,0);
imagesc(abs(tfrpf[end/2:-1:1,:]))
```
