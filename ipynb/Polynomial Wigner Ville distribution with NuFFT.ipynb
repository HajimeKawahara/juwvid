# Demonstration on the polynomial Wigner Ville with the non-uniform FFT

```julia
using PyPlot
import DSP
```

```julia
include("../juwvid.jl")
```

```julia
# multi component data
nsample=512
t,x=sampledata.genmultifm623x(nsample);
PyPlot.plot(t,x)
```

### polynomial WV with FFT

```julia
# the resolution is poor  
trfpo=polywv.tfrpowv(x,NaN,NaN,NaN,2,2);
a=juwplot.wtfrshow(abs(trfpo),t[2]-t[1],t[1],t[end],NaN,NaN,0.7*3,"Spectral")
PyPlot.ylim(0,40)
```

### with NuFFT

```julia
# now, the resolution is fine  
fin=collect(linspace(20,120,1024));
trfpo=polywv.tfrpowv(x,NaN,NaN,fin,2,2);
a=juwplot.wtfrshow(abs(trfpo),t[2]-t[1],t[1],t[end],NaN,NaN,0.7,"Spectral")
```
