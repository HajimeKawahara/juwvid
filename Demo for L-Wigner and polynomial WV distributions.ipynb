# Demo for the L-Wigner Distribution

```julia
using Winston;
using Color;
using PyPlot
import DSP
```

```julia
include("sampledata.jl")
include("jnufft.jl");
include("cohenclass.jl");
include("estif.jl");
include("juwutils.jl");
include("stft.jl");
include("smethod.jl");
include("lwigner.jl");
```

```julia
# multicomponent data (Boashash+15,p346,Example 6.2.3)
nsample=256
t,x=sampledata.genmultifm623(nsample);
Winston.plot(t,x)
```

```julia
# S-method (constant Lp)
sm=smethod.tfrsm(x,4,NaN,2)
imagesc(sm[end/2:-1:1,:])
```

```julia
#alias free sm
afwv=smethod.tfrsm(x,4,NaN,2)
#L2 wigner
trflw2=lwigner.trflw2L(afwv,4);
```

```julia
imagesc(trflw2[end/2:-1:1,:])
```
